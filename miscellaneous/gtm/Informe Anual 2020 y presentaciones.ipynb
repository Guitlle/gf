```python
def donutBLTarget(a, b, c, labels):
    plt.rcParams["figure.figsize"] = (5,5)
    fig, ax = plt.subplots()
    r = 0.80
    wedges, texts = ax.pie([0,100], radius=1, wedgeprops=dict(width=0.4), colors=["#aaaaaa55"])
    wedges, texts = ax.pie([a,100-a], radius=0.95, wedgeprops=dict(width=0.3), colors=["#ff6600bb", "#00000000"], startangle=90, )
    r = 0.80
    theta = 2 * np.pi * (a-4)/100
    rotation = 180+theta*180/(np.pi)
    if np.abs(rotation %360-180) < 90:
        rotation += 180
    plt.text(-np.sin(theta)*r, np.cos(theta)*r, str(a)+"%", color="white", rotation=rotation, 
             verticalalignment='center',
             horizontalalignment='center', fontsize=12)

    r=1.18
    theta = 2 * np.pi * (b)/100
    plt.plot([-np.sin(theta)*0.6, -np.sin(theta)*1], [np.cos(theta)*0.6, np.cos(theta)*1], color="black")
    rotation = 90+theta*180/(np.pi)
    if np.abs(rotation %360-180) < 90:
        rotation += 180
    plt.text(-np.sin(theta)*r, np.cos(theta)*r, str(b)+"%", color="black", rotation=rotation, 
             verticalalignment='center',
             horizontalalignment='center', fontsize=10)
    theta = 2 * np.pi * (c)/100
    plt.plot([-np.sin(theta)*0.6, -np.sin(theta)*1], [np.cos(theta)*0.6, np.cos(theta)*1], color="red")
    rotation = 90+theta*180/(np.pi)
    print(rotation)
    if np.abs(rotation %360-180) < 90:
        rotation += 180
    plt.text(-np.sin(theta)*r, np.cos(theta)*r, str(c)+"%", color="red", rotation=rotation,
             verticalalignment='center',
             horizontalalignment='center', fontsize=10)

    legend_elements = [
        mpl.lines.Line2D([0], [0], marker='o', linewidth=0, color='#ff6600bb', label=labels[0],
              markersize=10),
        mpl.lines.Line2D([0], [0], marker='o', linewidth=0, color='black', label=labels[1],
              markersize=10),
        mpl.lines.Line2D([0], [0], marker='o', linewidth=0, color='red', label=labels[2],
              markersize=10),]

    plt.legend(handles=legend_elements, loc='center', edgecolor="white")
```

```python
data = """	Extended Package	HIV+	Linked
CAS GUATEMALA	7241	290	204
FUNDAMACO	10340	218	118
SOMOS/APEVIHS	7661	45	20
CAS DEPARTMENTS	3262	78	57"""
data = [row.split("\t") for row in data.splitlines()]

datadf = pd.DataFrame(data).set_index(0).T.set_index('')
```

```python
datadf = datadf.astype(int)
```

```python
def human_format_int(num, a):
    num = int(num)
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%d%s' % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])

def human_format_float(num, a=None):
    num = float(num)
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return ('%.2f%s' if num%1>0 else '%d%s') % (num, ['', 'K', 'M', 'G', 'T', 'P'][magnitude])


plt.rcParams["grid.alpha"] = 0.4
plt.rcParams["figure.figsize"] = (8,3)
ax = datadf.astype(int).T.plot.bar(width=0.9, cmap="Paired")
plt.yscale("log")
plt.ylim(2, 2e4)
plt.xticks(rotation=0)
plt.xlabel("")
ax.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(human_format_int))
ax.yaxis.set_minor_formatter(mpl.ticker.NullFormatter())
plt.grid("on")
plt.legend(loc=(0.9,0.5))
j = 0
for i, row in datadf.T.iterrows():
    ax.text(j+0.05-1/3, int(row.values[0])*1.1, human_format_float(row.values[0]), horizontalalignment="center", color="#333333")
    ax.text(j, int(row.values[1])*1.1, human_format_float(row.values[1]), horizontalalignment="center", color="#333333")
    ax.text(j-0.05+1/3, int(row.values[2])*1.1, human_format_float(row.values[2]), horizontalalignment="center", color="#333333")
    j+=1
```

```python
#colores = mpl.cm.get_cmap("winter")
colores =  [ "#aaddee", "#6acd7d" , "#7aefca"] #colores([0.4,0.6,0.9])
ax = plt.subplot()
def plotNs(ns, offset = 1, alpha=1, color = "black", titles = None, orange=0.5, labelin = False, maxv=None, labelcolor="black"):
    n = len(ns)
    w = 360/n
    maxv = maxv if maxv is not None else np.max(ns) 
    for i,d in enumerate(ns):
        r = offset + d/maxv * orange
        a = ax.pie([0, 360-w, w], radius=r, startangle=i*w, wedgeprops=dict(width=r, linewidth=0.5), colors=[color, "#00000000"])
        for b in a[0][2:3]:
            b.set_edgecolor('black')
            b.set_alpha(alpha)
        r2 = r+ (-0.15 if labelin else 0.2)
        theta = (i-1.5)*w*np.pi/180
        rotation = 180+theta*180/(np.pi)
        if np.abs(rotation %360-180) < 90:
            rotation += 180
        plt.text(-np.sin(theta)*r2, np.cos(theta)*r2, str(d), color=labelcolor, rotation=rotation, 
                 verticalalignment='center',
                 horizontalalignment='center', fontsize=12)
        if titles is not None:
            r = r+0.7
            plt.text(-np.sin(theta)*r, np.cos(theta)*r, titles[i], color="black", rotation=0, 
                 verticalalignment='center',
                 horizontalalignment='center', fontsize=14)
plotNs(datadf.T["Extended Package"], 1.5, 1, colores[0], datadf.T.index.values, orange = 2, labelin=True)
plotNs(datadf.T["HIV+"], 0.35, 1, colores[1], orange = 1.2)
plotNs(datadf.T["Linked"], 0.35, 1, colores[2], orange = 1.2, labelin=True, maxv = 290)

legend_elements = [
        mpl.lines.Line2D([0], [0], marker='o', linewidth=0, color=colores[0], label="Extended package",
              markersize=10),
        mpl.lines.Line2D([0], [0], marker='o', linewidth=0, color=colores[1], label="HIV+",
              markersize=10),
        mpl.lines.Line2D([0], [0], marker='o', linewidth=0, color=colores[2], label="Linked",
              markersize=10),]

plt.legend(handles=legend_elements, loc=(0.1,1.2), edgecolor="white", prop={'size': 12})
```
