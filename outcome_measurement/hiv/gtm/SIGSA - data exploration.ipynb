```python
print(sigsa.groupby(["Orientación Sexual", "Año"]).apply(len).unstack(1).fillna(0))
def tempfun (x):
    cr =  (x["VIH-Resultado-c"]=="Reactivo").sum()
    cn =  (x["VIH-Resultado-c"]=="No Reactivo").sum()
    tr =  (x["VIH-Resultado-t"]=="Reactivo").sum()
    tn =  (x["VIH-Resultado-t"]=="No Reactivo").sum()
    return cr/(tr+tn)
print(sigsa.groupby(["Orientación Sexual", "Año"]).apply(tempfun).unstack(1).fillna(0))
```

```python
def tempfun (x):
    dn =  len(x[(x["VIH-Prueba de Tamizaje"]=="Si")]["Codigo de Persona"].unique())
    cr =  len(x[(x["VIH-Prueba de Tamizaje"]=="Si") & (x["VIH-Resultado-c"]=="Reactivo")]["Codigo de Persona"].unique())
    i = stats.beta.isf([0.025, 0.975], cr, dn-cr)
    if dn == 0:
        return "0"
    return str(round(100*cr/(dn), 2)) + " (" + str(round(i[1]*100, 2)) + " , " + str(round(i[0]*100, 2)) + ") [" + str(dn) + "]"

print("\n****************\Overall: ")
print(sigsa[sigsa["Año"]!=2014].groupby(["Año"]).apply(tempfun).fillna(0))
for var in ["Orientación Sexual", "Sexo", "Residencia Actual-Depto", "Pueblo", "Comunidad Liguistica", "Condición de Riesgo", "Motivo de Orientación", "Embarazo y Puerperio"]:
    print("\n****************\nVARIABLE: " + var)
    print(sigsa[sigsa["Año"]!=2014].groupby([var, "Año"]).apply(tempfun).unstack(1).fillna(0))
```
